```python
# First we import some useful modules
import csv # To read CSV files
from datetime import datetime # To get date objects from strings
from datetime import date # To make new date objects
import matplotlib.pyplot as plt # To graph
import matplotlib # To graph
```

```python
# Let's read our data file
data = []
dates = []
with open("res/SearchInterest.csv") as csvfile: # 'csvfile' is our file now
    reader = csv.reader(csvfile) # This Reader object allows us to iterate through rows
    for row in reader: # Extract the data row by row
        dates.append(matplotlib.dates.date2num(datetime.strptime(row[0], "%Y-%m-%d")))
        data.append(row[1])
```

```python
# Let's format the graph to have 'month' as X axis
fig, ax = plt.subplots(1)
fig.autofmt_xdate()
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%b"))

# Let's plot a vertical line for the date of the election
plt.plot_date([date(year=2016, month=11, day=6)]*2, [0, 100], '--',color='black', alpha=0.5)
# Let's plot our data
plt.plot_date(dates, data, "b-")

# Let's annotate the 
plt.annotate('pre-election', xy=(date(year=2016, month=8, day=20),20), alpha=0.5, style='italic', ha='center')
plt.annotate('post-election', xy=(date(year=2017, month=2, day=20),20), alpha=0.5, style='italic', ha='center')
plt.annotate(' "You\'re Fake News"\n     Trump to CNN ', xy=(date(year=2017, month=2, day=15),90), alpha=0.5, style='italic', ha='left')
plt.annotate(' ', xy=(date(year=2017, month=1, day=10),97),
             xytext=(date(year=2017, month=2, day=15),97), 
             alpha=0.5, style='italic', ha='left',
             arrowprops=dict(arrowstyle='->', alpha=0.5))
plt.title("Fake News")
plt.ylabel("Weekly Google Search Interest (%)")
plt.xlabel("Date")
plt.show()
```
